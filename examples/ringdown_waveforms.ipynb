```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import fdringdown as rd
```

```python
wf = rd.waveforms.ringdown_v2(modes=[(2,2,0),(3,3,0)])
```

```python
times = np.linspace(-0.01, 0.05, num=200)
params = {
    'A_rd_0': 1.,
    'A_rd_1': 0.5,
    'phi_rd_0': 0,
    'phi_rd_1': 0,
    'iota': np.pi/6,
    'varphi': 0,
    't_0': 0,
    'M_f': 70,
    'chi_f': 0.7
}
```

```python
pyring_wf = wf.waveform(times, params)
```

```python
fig, ax = plt.subplots()

ax.plot(times, np.real(pyring_wf))
ax.plot(times, -np.imag(pyring_wf))
```

```python
isi_wf = wf.waveform_isi(times, params)
```

```python
fig, ax = plt.subplots()

ax.plot(times, np.real(isi_wf))
ax.plot(times, -np.imag(isi_wf))
```

```python
fig, ax = plt.subplots(dpi=180)

ax.plot(times, np.real(pyring_wf))
ax.plot(times, np.real(isi_wf))
```

```python
fig, ax = plt.subplots(dpi=180)

ax.plot(times, np.imag(pyring_wf))
ax.plot(times, np.imag(isi_wf))
```
